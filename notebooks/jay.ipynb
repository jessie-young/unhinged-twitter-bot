```python
import daft
from openai import OpenAI
import instructor
import os
import pydantic_to_pyarrow
import pydantic
import pyarrow as pa

def pydantic_udf(pydantic_model: pydantic.BaseModel, udf: type, **kwargs):
    pyarrow_schema = pydantic_to_pyarrow.get_pyarrow_schema(pydantic_model)
    pyarrow_dtype = pa.struct([(c, pyarrow_schema.field(c).type) for c in pyarrow_schema.names])
    decorator = daft.udf(return_dtype=daft.DataType.from_arrow_type(pyarrow_dtype), **kwargs)
    return decorator(udf)

class StructuredLLM:
    def __init__(self):
        self.client = instructor.from_openai(OpenAI(api_key=os.getenv("OPENAI_API_KEY")))

    def __call__(self, prompt, response_model):
        data = prompt.to_pylist()
        results = []
        for p in data:
            response = self.client.chat.completions.create(
                model="gpt-3.5-turbo",
                response_model=response_model,
                messages=[
                    # {
                    #     "role": "system",
                    #     "content": "You are an AI agent responding to tweets. Think through your response step by step.",
                    # },
                    {"role": "user", "content": p},
                ],
                temperature=0.7,
            )
            results.append(response)
        return [r.model_dump() for r in results]
        
```

```python
import pydantic

class CringeLevel(pydantic.BaseModel):
    score: float
    reason: str

fast_structured_llm = pydantic_udf(CringeLevel, StructuredLLM, batch_size=1).with_concurrency(8)

df = df.with_column("cringe_level", fast_structured_llm("Rate this tweet on level of cringeness, from 0 to 1 and provide a concise reason as to why:\n\n" + df["response"], CringeLevel))
```

```python
df.collect()
```

```python
df = df.with_column("cringe_score", df["cringe_level"].struct.get("score"))
df = df.with_column("cringe_reason", df["cringe_level"].struct.get("reason"))
```
