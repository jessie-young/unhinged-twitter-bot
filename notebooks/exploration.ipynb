```python
import daft
```

```python
df = daft.read_json("../data/agent-logs/simulation-1/**")
df.show()
```

```python
df = df.with_column("prompt", df["data"].struct.get("prompt"))
df = df.with_column("response", df["data"].struct.get("response"))
df = df.exclude("data")
df.show()
```

```python
df.where(df["prompt_type"] == "tweet_generation").select("session_id", "response").show()
```
